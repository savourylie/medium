```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as display
import time
import seaborn as sns
%matplotlib inline
```

```python
mean1 = [0, 0]
mean2 = [10, 10]
cov1 = [[100, 0], [0, 100]]
cov2 = [[30, 0], [0, 30]]

a, b = np.random.multivariate_normal(mean1, cov1, 5000).T
x, y = np.random.multivariate_normal(mean2, cov2, 5000).T
x1 = np.concatenate((a, x))
x2 = np.concatenate((b, y))
label = np.array([1] * 5000 + [0] * 5000)

data = pd.DataFrame({'x': x1, 'y': x2, 'z': label})

sns.lmplot('x', # Horizontal axis
           'y', # Vertical axis
           data=data, # Data source
           fit_reg=False, # Don't fix a regression line
           hue="z", # Set color
           scatter_kws={"marker": "D", # Set marker style
                        "s": 50, "alpha": 0.3}) # S marker size
```

```python
features, label = data.drop(['z'], axis=1), data['z']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=1113)
```

```python
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.metrics import silhouette_score
import operator

scores = {}

for n in range(2, 14):
    clusterer = MiniBatchKMeans(init='k-means++', n_clusters=n, batch_size=512, n_init=10, max_no_improvement=10, verbose=0, random_state=1113)
    # clusterer = KMeans(n_clusters = n)
    clusterer.fit(X_train)
    
    preds = clusterer.predict(X_train)
    centers = clusterer.cluster_centers_
    cluster_labels = clusterer.labels_
    score = silhouette_score(X_train, cluster_labels, metric='euclidean', sample_size=2048, random_state=1113)
    scores[n] = score
    
    print("Silhouette score with n = {}: {}".format(n, score))

max_key = max(scores.keys(), key=(lambda k: scores[k])) 
print(max_key)
```

```python
clusterer = MiniBatchKMeans(init='k-means++', n_clusters=max_key, batch_size=512, n_init=10, max_no_improvement=10, verbose=0, random_state=1113)
clusters_train = clusterer.fit_predict(X_train)
clusters_test = clusterer.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

clusters_test = [1 if x == 0 else 0 for x in clusters_test]

print(max(roc_auc_score(y_test, clusters_test), 1 - roc_auc_score(y_test, clusters_test)))
```
